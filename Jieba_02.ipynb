```python
import jieba
import glob
```

```python
glob.glob('./res_gossiping/*.txt')
```

```python
sb = ""
for filepath in glob.glob('./res_gossiping/*.txt'):
    sb += open(filepath, 'r', encoding='utf-8') \
            .read() \
            .split('---split---')[0]
sb
```

```python
wc_generator = jieba.cut(sb)
```

```python
wc_list = []

for w in wc_generator:
#     print(w)
    wc_list.append(w)
wc_list
```

```python
wc_list.count('作者')
```

```python
# 第一種

wc_map = {}
for w in set(wc_list):
    wc_map[w] = wc_list.count(w)
    
wc_map
```

```python
# 第二種，較有效率

wc_map = {}
for w in wc_list:
    if w in wc_map:
        wc_map[w] += 1
    else:
        wc_map[w] = 1
        
wc_map
```

```python
wc_sort_list = [(k, v) for k, v in wc_map.items()]          # dict無法進行排序，先轉成元組類型
wc_sort_list
```

```python
wc_sort_list.sort(key=lambda x: x[1], reverse=True)                 # key裡設匿名函數lambda x: x[1]-->取元組的第二個值進行降冪排列
wc_sort_list
```

```python
stopwords = open('stopword.txt', 'r', encoding='utf-8').readlines()
stopwords = [i.replace("\n", "") for i in stopwords]
stopwords.append("\n")

sb = ""
for filepath in glob.glob('./res_gossiping/*.txt'):
    sb += open(filepath, 'r', encoding='utf-8') \
            .read() \
            .split('---split---')[0]
    
wc_generator = jieba.cut(sb)

wc_list = []

for w in wc_generator:
#     print(w)
    wc_list.append(w)

wc_map = {}

for w in wc_list:
    if w in stopwords or w.__len__() < 2:
        continue
    if w in wc_map:
        wc_map[w] += 1
    else:
        wc_map[w] = 1

wc_sort_list = [(k, v) for k, v in wc_map.items()]  

wc_sort_list.sort(key=lambda x: x[1], reverse=True)                 
wc_sort_list
```
