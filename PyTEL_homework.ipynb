```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import os
from openpyxl import Workbook


if not os.path.exists('./job_homework'):
    os.mkdir('./job_homework')    


url = "https://www.104.com.tw/jobs/search/list?ro=0&jobcat=2007001000&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&area=6001005008%2C6001002023&order=11&asc=0&page=2&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1"

# res = requests.get(url)
useragent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"
Referers = "https://www.104.com.tw/jobs/search/?ro=0&jobcat=2007001000&expansionType=area,spec,com,job,wf,wktm&area=6001005008,6001002023&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1"
headerstr = {"User-Agent": useragent, "Referer": Referers}

res = requests.get(url, headers=headerstr)
# resp = json.loads(res.text)

```

```python
df = pd.DataFrame(res.json()['data']['list'])
df
```

```python
df = df.drop(['jobType', 
              'jobNo', 
              'jobNameSnippet', 
              'jobRole', 
              'jobRo', 
              'jobAddrNo', 
              'tags', 
              'landmark', 
              'link', 
              'jobsource', 
              'jobNameRaw', 
              'lon', 
              'lat', 
              'remoteWorkType', 
               ], axis=1)
df
```

```python
df1 = df.drop([
    'period', 
    'periodDesc', 
    'applyCnt', 
    'applyDesc', 
    'custNo', 
    's10', 
    'appearDate', 
    'appearDateDesc', 
    'optionZone', 
    'isApply', 
    'applyDate', 
    'isSave',
    'coIndustry', 
    'salaryLow', 
    'salaryHigh', 
    'major',
    'custNameRaw',
    'descSnippet'], axis=1)
df1
```

```python
df1.columns
```

```python
df1['公司地址'] = df1['jobAddrNoDesc'] + df1['jobAddress']
df1 = df1.drop(['jobAddrNoDesc', 'jobAddress'], axis=1)
df1
```

```python
df1.rename(columns = {'jobName':'職位名稱', 
                     'description':'工作內容', 
                     'optionEdu':'學歷', 
                     'custName':'公司名稱',
                     'coIndustryDesc':'行業類別',
                     'salaryDesc':'薪資待遇'}, inplace=True)
```

```python
df1.to_excel('./job_homework/job104.xlsx', encoding="utf-8")
```
